```python
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
```

```python
iris
```

```python
sns.set(style='ticks')
```

```python
sns.pairplot(iris)
```

```python
sns.pairplot(iris,hue='species')
```

```python
sns.pairplot(iris,hue='species',palette='husl')
```

```python
sns.pairplot(iris,hue='species',palette='Blues_d')
```

```python
sns.pairplot(iris,hue='species',vars=['sepal_length','sepal_width'],size=3)
```

```python
sns.pairplot(iris,hue='species',x_vars=['sepal_length','sepal_width'],y_vars=['petal_length','petal_width'])
```
